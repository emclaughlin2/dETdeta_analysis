```python
def divide_histograms(h1, h2):
    h3 = TH1F(h1.Clone("h3"))
    h3.Divide(h2)

    for i in range(1, h3.GetNbinsX() + 1):
        val1 = h1.GetBinContent(i)
        err1 = h1.GetBinError(i)
        val2 = h2.GetBinContent(i)
        err2 = h2.GetBinError(i)

        # Calculate error for h3 using error propagation formula
        val3 = h3.GetBinContent(i)
        err3 = 0.0
        if val2 != 0:
            err3 = abs(val3) * ((err1 / val1)**2 + (err2 / val2)**2)**0.5
        h3.SetBinError(i, err3)

    return h3
```
