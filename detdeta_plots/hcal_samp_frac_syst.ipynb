```python
def fit_histogram_to_gaussian(hist, minusx, plusx):
    # Define the Gaussian function
    gaussian_fit = ROOT.TF1("gaussian_fit", "gaus", hist.GetMean()-minusx, hist.GetMean()+plusx)

    # Set initial parameter values (you can adjust these based on your data)
    mean_initial = hist.GetMean()
    sigma_initial = hist.GetRMS()
    amplitude_initial = hist.GetMaximum()

    gaussian_fit.SetParameters(amplitude_initial, mean_initial, sigma_initial)

    # Perform the fit
    fit_result = hist.Fit(gaussian_fit, "RS")

    # Check if the fit is successful
    if fit_result.IsValid():
        print("Fit successful!")
        print(gaussian_fit.GetParameter(1), gaussian_fit.GetParameter(2))
        return gaussian_fit.GetParameter(1), gaussian_fit.GetParError(1), gaussian_fit.GetParameter(2), gaussian_fit.GetParError(2)
    else:
        print("Fit failed!")
        return None, None, None, None
```
