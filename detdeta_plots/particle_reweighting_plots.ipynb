```python
fin = ROOT.TFile("scaleEPOS.root")

postfix = ["0010_ratio", "1020_ratio", "2040_ratio", "4060_ratio", "6092_ratio"]

h_pimi = [None] * 5
h_pip = [None] * 5
h_p = [None] * 5
h_pbar = [None] * 5
h_kp = [None] * 5
h_kmi = [None] * 5

# Retrieve histograms
for i in range(5):
    h_pimi[i] = fin.Get("h_pimi{}".format(postfix[i]))
    h_pip[i] = fin.Get("h_pip{}".format(postfix[i]))
    h_p[i] = fin.Get("h_p{}".format(postfix[i]))
    h_pbar[i] = fin.Get("h_pbar{}".format(postfix[i]))
    h_kp[i] = fin.Get("h_kp{}".format(postfix[i]))
    h_kmi[i] = fin.Get("h_kmi{}".format(postfix[i]))

# Fit each histogram with a Padé approximant 2/2
f_pimi = [None] * 5
f_pip = [None] * 5
f_p = [None] * 5
f_pbar = [None] * 5
f_kp = [None] * 5
f_kmi = [None] * 5

for i in range(5):
    formula = "([0]+[1]*x+[2]*x*x) / (1+[3]*x+[4]*x*x)"
    f_pimi[i] = ROOT.TF1("f_pimi{}".format(postfix[i]), formula, h_pimi[i].GetXaxis().GetXmin(), h_pimi[i].GetXaxis().GetXmax())
    f_pip[i] = ROOT.TF1("f_pip{}".format(postfix[i]), formula, h_pip[i].GetXaxis().GetXmin(), h_pip[i].GetXaxis().GetXmax())
    f_p[i] = ROOT.TF1("f_p{}".format(postfix[i]), formula, h_p[i].GetXaxis().GetXmin(), h_p[i].GetXaxis().GetXmax())
    f_pbar[i] = ROOT.TF1("f_pbar{}".format(postfix[i]), formula, h_pbar[i].GetXaxis().GetXmin(), h_pbar[i].GetXaxis().GetXmax())
    f_kp[i] = ROOT.TF1("f_kp{}".format(postfix[i]), formula, h_kp[i].GetXaxis().GetXmin(), h_kp[i].GetXaxis().GetXmax())
    f_kmi[i] = ROOT.TF1("f_kmi{}".format(postfix[i]), formula, h_kmi[i].GetXaxis().GetXmin(), h_kmi[i].GetXaxis().GetXmax())

    # Example parameter setting (commented out)
    #f_p[i].SetParameters(5.18368e-01,  -1.5e-01, 1.86998e-02,  -3.67333e-01, 4.0e-02);
    #f_pbar[i].SetParameters(1,-1,1,0,0);
    #f_kmi[i].SetParameters(1, -1, 1, 1, 0);
    
    #epos
    f_p[i].SetParameters(5.18368e-01,  -1.5e-01, 1.86998e-02,  -3.67333e-01, 4.0e-02);
    f_pbar[i].SetParameters(7.18399e-01,-2.68459e-01,-2.70367e-02,-3.61821e-01,3.38751e-02);
    f_kmi[i].SetParameters(1, -1, 1, 1, 0);

    h_pimi[i].Fit("f_pimi{}".format(postfix[i]), "REM")
    h_pip[i].Fit("f_pip{}".format(postfix[i]), "REM")
    h_p[i].Fit("f_p{}".format(postfix[i]), "REML")
    h_pbar[i].Fit("f_pbar{}".format(postfix[i]), "REM")
    h_kp[i].Fit("f_kp{}".format(postfix[i]), "REM")
    h_kmi[i].Fit("f_kmi{}".format(postfix[i]), "REM")
```

```python
fin = ROOT.TFile("EPOSLambdaRatio.root")

# Initialize arrays for histograms and fit functions
lambdaratio = [None] * 5
lambdabaratio = [None] * 5
f_lambdaratio = [None] * 5
f_lambdabaratio = [None] * 5

# Retrieve histograms and define fit functions
for i in range(5):
    lambdaratio[i] = fin.Get(f"lambdaratio_{i}")
    lambdabaratio[i] = fin.Get(f"lambdabaratio_{i}")

    f_lambdaratio[i] = ROOT.TF1(f"f_lambdaratio_{i}", "([0]+[1]*x+[2]*x*x) / (1+[3]*x+[4]*x*x)", lambdaratio[i].GetXaxis().GetXmin(), lambdaratio[i].GetXaxis().GetXmax())
    f_lambdabaratio[i] = ROOT.TF1(f"f_lambdabarratio_{i}", "([0]+[1]*x+[2]*x*x) / (1+[3]*x+[4]*x*x)", lambdabaratio[i].GetXaxis().GetXmin(), lambdabaratio[i].GetXaxis().GetXmax())

    #EPOS
    f_lambdaratio[i].SetParameters( -2.02072e+03, 1.31847e+04, -2.09487e+03, 7.56991e+03, -9.75871e+02);
    f_lambdabaratio[i].SetParameters( -2.02072e+03, 1.31847e+04, -2.09487e+03, 7.56991e+03, -9.75871e+02);
    
    lambdaratio[i].Fit(f"f_lambdaratio_{i}", "REMI")
    lambdabaratio[i].Fit(f"f_lambdabarratio_{i}", "REMI")
```
