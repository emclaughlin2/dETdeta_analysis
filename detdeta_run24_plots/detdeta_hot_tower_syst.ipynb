```python
files = ["test_intermittant_ht/dETdeta_variation_tprofile_ht_0-1.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_1-2.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_2-3.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_3-4.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_4-5.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_55-56.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_56-57.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_57-58.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_58-59.root",
"test_intermittant_ht/dETdeta_variation_tprofile_ht_59-60.root"]

cents = ['0-1','1-2','2-3','3-4','4-5','55-56','56-57','57-58','58-59','59-60']

h_emcal = []
h_ihcal = []
h_ohcal = []
h_calo = []
h_hcal = []

for i, file in enumerate(files):
    f1 = ROOT.TFile.Open(file)
    h_emcal.append(TH1F(f1.Get("emcal_reco_dev")))
    h_ihcal.append(TH1F(f1.Get("ihcal_reco_dev")))
    h_ohcal.append(TH1F(f1.Get("ohcal_reco_dev")))
    h_calo.append(TH1F(f1.Get("calo_reco_dev")))
    h_hcal.append(TH1F(f1.Get("hcal_reco_dev")))
    h_emcal[i].SetDirectory(0)
    h_ihcal[i].SetDirectory(0)
    h_ohcal[i].SetDirectory(0)
    h_calo[i].SetDirectory(0)
    h_hcal[i].SetDirectory(0)
    f1.Close()
```

```python
for i in range(len(h_emcal)):
    canvas = TCanvas("canvas","",600,500)
    canvas.SetRightMargin(0.1)
    leg = ROOT.TLegend(.4,.8,.8,.94)
    leg.SetTextSize(0.04)
    leg.SetFillStyle(0)
    leg.AddEntry("","#it{#bf{sPHENIX}} Internal","")
    leg.AddEntry("","Run 54912 ana450_2024p009","")
    h_emcal[i].SetLineColor(1)
    h_emcal[i].SetMarkerColor(1)
    leg.AddEntry(h_emcal[i],cents[i]+'%',"lep")
    h_emcal[i].GetXaxis().SetTitleSize(0.04)
    h_emcal[i].GetYaxis().SetTitleSize(0.04)
    h_emcal[i].GetXaxis().SetLabelSize(0.04)
    h_emcal[i].GetYaxis().SetLabelSize(0.04)
    h_emcal[i].GetYaxis().SetTitleOffset(1.8)
    h_emcal[i].SetXTitle("EMCal i#eta")
    h_emcal[i].SetYTitle("EMCal dE_{T}/d#eta variation [GeV]")
    h_emcal[i].GetYaxis().SetRangeUser(-0.015,0.015)
    h_emcal[i].Draw()
    leg.Draw()
    canvas.Draw()
    canvas.SaveAs('/sphenix/u/egm2153/'+direct+'/emcal_hot_tower_mask_'+cents[i]+'.png')
```
